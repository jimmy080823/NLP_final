```python
import pandas as pd
import numpy as np
import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec

# 確保停用詞已經被下載
nltk.download('stopwords')

column_names = ['label', 'text']
data = pd.read_csv('SpamVsHam.tsv', delimiter='\t', header=None, names=column_names)

# 文本清理函數
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\b\w*\d\w*\b', '', text)  # 修改這裡，只移除完全是單詞的數字
    return text

# 清理文本數據
data['text'] = data['text'].apply(clean_text)

# 停用詞處理
stop_words = set(stopwords.words('english') + ['u', 'im', 'c'])
data['text'] = data['text'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop_words))

# 詞幹處理
stemmer = SnowballStemmer("english")
data['text'] = data['text'].apply(lambda x: ' '.join(stemmer.stem(word) for word in x.split()))

# Basic text preprocessing
# Convert text to lowercase
data['text'] = data['text'].str.lower()

# Check for missing values
print(data.isnull().sum())

# Handle missing values if there are any
data = data.dropna()

def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

data['text'] = data['text'].apply(remove_emoji)

def remove_url(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub('', text)

data['text'] = data['text'].apply(remove_url)

def remove_html(text):
    html_pattern = re.compile(r'<.*?>')
    return html_pattern.sub('', text)

data['text'] = data['text'].apply(remove_html)

# Check the balance of the classes
print(data['label'].value_counts())

# 展示一些處理後的樣本
print(data.sample(10))

```

#  TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
tfidf_vectors = tfidf_vectorizer.fit_transform(data['text'])

# Convert to dense array and show example
tfidf_vectors_dense = tfidf_vectors.todense()
print(tfidf_vectors_dense[0:10])  # Show first 10 vector representations

print("Length of each document's vector:", tfidf_vectors.shape)
```

# Word2Vec

```python
from gensim.models import Word2Vec
import nltk

# Tokenize the text
data['tokens'] = data['text'].apply(nltk.word_tokenize)

# Create and train the Word2Vec model
w2v_model = Word2Vec(sentences=data['tokens'], vector_size=100, window=5, min_count=1, workers=4)

# Generate the average Word2Vec for each document
def document_vector(doc):
    # Remove out-of-vocabulary words
    doc = [word for word in doc if word in w2v_model.wv.key_to_index]
    if not doc:
        return np.zeros(w2v_model.vector_size)
    return np.mean(w2v_model.wv[doc], axis=0)

# Apply the function to each document
data['doc_vector'] = data['tokens'].apply(document_vector)

# Show example vectors
print(data['doc_vector'].head(10))
if not data['doc_vector'].empty:
    print("Shape of Word2Vec vectors:", len(data['doc_vector'][0]), "x 1")
else:
    print("No Word2Vec vectors available.")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test_tfidf, y_pred_tfidf)
print("Confusion Matrix:\n", conf_matrix)

# Function to plot the confusion matrix
def plot_conf_matrix(cm, classes,
                     normalize=False,
                     title='Confusion matrix',
                     cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], '.2f' if normalize else 'd'),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

# Call the function to plot the matrix
plot_conf_matrix(conf_matrix, classes=['HAM', 'SPAM'], title='Confusion Matrix', normalize=False)
plt.show()
```

# Deep learning

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Assuming data loading and preparation is done here...

# Parameters
input_dim_tfidf = X_train_tfidf.shape[1]  # Number of TF-IDF features
input_dim_w2v = X_train_w2v.shape[1]      # Number of Word2Vec features
hidden_dim = 128
output_dim = 1
batch_size = 32
epochs = 5

# Label encoding
label_encoder = LabelEncoder()
y_encoded_tfidf = label_encoder.fit_transform(y_train_tfidf)
y_encoded_w2v = label_encoder.fit_transform(y_train_w2v)

# Splitting the data into training and validation sets for both feature sets
X_train_tfidf, X_val_tfidf, y_train_tfidf, y_val_tfidf = train_test_split(
    X_train_tfidf, y_encoded_tfidf, test_size=0.2, random_state=42)
X_train_w2v, X_val_w2v, y_train_w2v, y_val_w2v = train_test_split(
    X_train_w2v, y_encoded_w2v, test_size=0.2, random_state=42)

# Convert to tensors
X_train_tfidf_tensor = torch.FloatTensor(X_train_tfidf.toarray())
y_train_tfidf_tensor = torch.FloatTensor(y_train_tfidf).unsqueeze(1)
X_val_tfidf_tensor = torch.FloatTensor(X_val_tfidf.toarray())
y_val_tfidf_tensor = torch.FloatTensor(y_val_tfidf).unsqueeze(1)

X_train_w2v_tensor = torch.FloatTensor(X_train_w2v)
y_train_w2v_tensor = torch.FloatTensor(y_train_w2v).unsqueeze(1)
X_val_w2v_tensor = torch.FloatTensor(X_val_w2v)
y_val_w2v_tensor = torch.FloatTensor(y_val_w2v).unsqueeze(1)

# Creating datasets and dataloaders
train_data_tfidf = TensorDataset(X_train_tfidf_tensor, y_train_tfidf_tensor)
train_loader_tfidf = DataLoader(train_data_tfidf, batch_size=batch_size, shuffle=True)
val_data_tfidf = TensorDataset(X_val_tfidf_tensor, y_val_tfidf_tensor)
val_loader_tfidf = DataLoader(val_data_tfidf, batch_size=batch_size, shuffle=False)

train_data_w2v = TensorDataset(X_train_w2v_tensor, y_train_w2v_tensor)
train_loader_w2v = DataLoader(train_data_w2v, batch_size=batch_size, shuffle=True)
val_data_w2v = TensorDataset(X_val_w2v_tensor, y_val_w2v_tensor)
val_loader_w2v = DataLoader(val_data_w2v, batch_size=batch_size, shuffle=False)

# Define the model class as before
class DenseClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(DenseClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return torch.sigmoid(x)

# Define and instantiate the model for TF-IDF
model_tfidf = DenseClassifier(input_dim_tfidf, hidden_dim, output_dim)
optimizer_tfidf = optim.Adam(model_tfidf.parameters(), lr=0.001)

# Define and instantiate the model for Word2Vec
model_w2v = DenseClassifier(input_dim_w2v, hidden_dim, output_dim)
optimizer_w2v = optim.Adam(model_w2v.parameters(), lr=0.001)

# Function to train the model
def train_and_validate_model(model, train_loader, val_loader, optimizer, criterion, epochs, device):
    model.to(device)
    for epoch in range(epochs):
        model.train()
        total_train_loss = 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            total_train_loss += loss.item()
        
        # Validation loss
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                total_val_loss += loss.item()

        print(f'Epoch {epoch+1}/{epochs}, Train Loss: {total_train_loss/len(train_loader):.4f}, Validation Loss: {total_val_loss/len(val_loader):.4f}')

# Define the loss function
criterion = nn.BCELoss()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Training the model for TF-IDF
print("Training the Dense model for TF-IDF features...")
train_and_validate_model(model_tfidf, train_loader_tfidf, val_loader_tfidf, optimizer_tfidf, criterion, epochs, device)

# Training the model for Word2Vec
print("\nTraining the Dense model for Word2Vec features...")
train_and_validate_model(model_w2v, train_loader_w2v, val_loader_w2v, optimizer_w2v, criterion, epochs, device)
```

# CNN

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

class CNNClassifier(nn.Module):
    def __init__(self, input_dim):
        super(CNNClassifier, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)  # Convolutional layer
        self.relu = nn.ReLU()  # Activation function
        self.maxpool = nn.MaxPool1d(kernel_size=2)  # Max pooling layer
        self.dropout = nn.Dropout(0.5)  # Dropout to prevent overfitting
        self.fc = nn.Linear(32 * (input_dim // 2), 1)  # Fully connected layer

    def forward(self, x):
        x = x.unsqueeze(1)  # Add a channel dimension
        x = self.conv1(x)
        x = self.relu(x)
        x = self.maxpool(x)
        x = x.view(x.size(0), -1)  # Flatten the output for the fully connected layer
        x = self.dropout(x)
        x = self.fc(x)
        return torch.sigmoid(x)

# Parameters
input_dim_tfidf = X_train_tfidf.shape[1]  # Number of TF-IDF features
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Model, Loss, and Optimizer
model_tfidf = CNNClassifier(input_dim_tfidf).to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model_tfidf.parameters(), lr=0.001)

# DataLoader
X_train_tensor = torch.FloatTensor(X_train_tfidf.toarray()).to(device)
y_train_tensor = torch.FloatTensor(y_train_tfidf).unsqueeze(1).to(device)
dataset = TensorDataset(X_train_tensor, y_train_tensor)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

# Training Function
def trainCNN(model, loader, optimizer, criterion, epochs):
    model.train()
    for epoch in range(epochs):
        for inputs, labels in loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        print(f'Epoch {epoch+1}: Loss = {loss.item()}')

# Train the model
trainCNN(model_tfidf, loader, optimizer, criterion, epochs=5)
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import csr_matrix 
from sklearn.metrics import accuracy_score

# Convert labels to numeric values
label_encoder = LabelEncoder()
y_train_tfidf = label_encoder.fit_transform(y_train_tfidf)
y_val_tfidf = label_encoder.transform(y_val_tfidf)
y_train_w2v = label_encoder.transform(y_train_w2v)
y_val_w2v = label_encoder.transform(y_val_w2v)

class LSTMDataset(Dataset):
    def __init__(self, X, y):
        if isinstance(X, csr_matrix):
            X = X.toarray()
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        data = self.X[idx].unsqueeze(1)   # Add sequence length dimension
        label = self.y[idx]
        return data, label

# Dataset and DataLoader setup
BATCH_SIZE = 32
train_dataset_tfidf = LSTMDataset(X_train_tfidf, y_train_tfidf)
val_dataset_tfidf = LSTMDataset(X_val_tfidf, y_val_tfidf)
train_loader_tfidf = DataLoader(train_dataset_tfidf, batch_size=BATCH_SIZE, shuffle=True)
val_loader_tfidf = DataLoader(val_dataset_tfidf, batch_size=BATCH_SIZE)

train_dataset_w2v = LSTMDataset(X_train_w2v, y_train_w2v)
val_dataset_w2v = LSTMDataset(X_val_w2v, y_val_w2v)
train_loader_w2v = DataLoader(train_dataset_w2v, batch_size=BATCH_SIZE, shuffle=True)
val_loader_w2v = DataLoader(val_dataset_w2v, batch_size=BATCH_SIZE)

class TEXT_LSTM(nn.Module):
    def __init__(self, input_size, hidden_size=64, num_classes=1):
        super(TEXT_LSTM, self).__init__()
        self.rnn = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=1, bidirectional=True, batch_first=True)
        self.drop = nn.Dropout(p=0.5)
        self.fc = nn.Linear(2 * hidden_size, num_classes)

    def forward(self, X_batch):
        output, (hidden, cell) = self.rnn(X_batch)
        text_fea = self.drop(output[:, -1])
        text_fea = self.fc(text_fea)
        text_out = torch.sigmoid(text_fea).squeeze()
        return text_out

# Define and train models
def trainLSTM(model, optimizer, criterion, train_loader, valid_loader, num_epochs, eval_every, file_path, best_valid_loss=float('Inf')):
    running_loss, valid_running_loss = 0.0, 0.0
    global_step, train_loss_list, valid_loss_list, global_steps_list = 0, [], [], []

    model.train()
    for epoch in range(num_epochs):
        for text, labels in train_loader:
            text, labels = text.to(device), labels.to(device).float()
            output = model(text)
            loss = criterion(output, labels)
            running_loss += loss.item()

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            global_step += 1
            if global_step % eval_every == 0:
                model.eval()
                with torch.no_grad():
                    val_labels_all = []
                    val_preds_all = []
                    for text, labels in valid_loader:
                        text, labels = text.to(device), labels.to(device).float()
                        output = model(text)
                        loss = criterion(output, labels)
                        valid_running_loss += loss.item()

                        val_preds = torch.round(output).cpu().detach().numpy()
                        val_labels = labels.cpu().detach().numpy()

                        val_labels_all.extend(val_labels)
                        val_preds_all.extend(val_preds)

                    average_train_loss = running_loss / eval_every
                    average_valid_loss = valid_running_loss / len(valid_loader)
                    train_loss_list.append(average_train_loss)
                    valid_loss_list.append(average_valid_loss)
                    global_steps_list.append(global_step)

                    # Calculate accuracy
                    val_accuracy = accuracy_score(val_labels_all, val_preds_all)

                    print(f'Epoch[{epoch+1}/{num_epochs}], Step[{global_step}], Train Loss: {average_train_loss:.4f}, Valid Loss: {average_valid_loss:.4f}, Valid Accuracy: {val_accuracy:.4f}')
                    running_loss, valid_running_loss = 0.0, 0.0
                    model.train()

    print('Training complete.')

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Training on device: {device}')

# Instantiate models and start training
input_size_tfidf = X_train_tfidf.shape[1]
model_tfidf = TEXT_LSTM(input_size_tfidf).to(device)
optimizer_tfidf = optim.Adam(model_tfidf.parameters(), lr=1e-3)
criterion = nn.BCELoss()

print("Training with TF-IDF features:")
trainLSTM(model_tfidf, optimizer_tfidf, criterion, train_loader_tfidf, val_loader_tfidf, 5, len(train_loader_tfidf) // 2, "output/tfidf")

input_size_w2v = X_train_w2v.shape[1]
model_w2v = TEXT_LSTM(input_size_w2v).to(device)
optimizer_w2v = optim.Adam(model_w2v.parameters(), lr=1e-3)

print("\nTraining with Word2Vec features:")
trainLSTM(model_w2v, optimizer_w2v, criterion, train_loader_w2v, val_loader_w2v, 5, len(train_loader_w2v) // 2, "output/w2v")
```
